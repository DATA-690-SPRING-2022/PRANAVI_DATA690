# Cleansing the data to make the year, population, cost, cost per person numerical type - use apply() function

# census year

```python
df["cleansed_year"]=df["Census Year"].apply(lambda x : x.replace("*",""))
df["cleansed_year"]
```

# population

```python
df["cleansed_pop"]=df["Total Population"].apply(lambda x:x.replace(",",""))
df["cleansed_pop"]
```

# cost

```python
df["cleansed_cost"]=df["Census Cost"].apply(lambda x:x.replace(",","").replace("$","").replace(" ","").replace("Billion","000000000000").replace(".50","5"))
df["cleansed_cost"]
```

# cost per person

```python
df["cleansed_avg"] = df["Average Cost Per Person"].apply(lambda x:x.replace('cents','').replace("$",""))
df['cleansed_avg']
```

```python
df3=df[df.columns[4:]]
df3
```

```python
df3["cleansed_cost"]=df3["cleansed_cost"].astype(int)
df3["cleansed_year"]=df3["cleansed_year"].astype(int)
df3["cleansed_pop"]=df3["cleansed_pop"].astype(int)
df3["cleansed_avg"]=df3["cleansed_avg"].astype(float)
df3
```

```python
df3.info()
```

# Use Pandas built-in plotting function to plot the average cost per person over time.

```python
df3.plot(x ='cleansed_year', y='cleansed_avg',xlabel = ' Year ', ylabel = 'Avg Cost per Person' )
```
