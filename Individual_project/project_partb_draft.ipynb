```python
df = pd.read_csv(DATA_URL)

df.sample(10)
```

```python
df_2020_pop = df.query("Year == 2010").query("indicator == 'BAR.NOED.1519.ZS'")
df_2020_pop.sample(5)
```

```python
df_2020_pop_sort = df_2020_pop.sort_values(by="value", ascending=False)

df_2020_pop_sort.head()
```

```python
fig = px.bar(
    df_2020_pop_sort, 
    x="Country Name", labels={"value":"2010 Population Barro-Lee:Percentage of population age 15-19 with no education", "Country Name":"Country Name"},
    y="value",
    color="Country Name",
    height=800,
    
)

fig.update_layout(showlegend=False)

fig.show()
```

Here in the bar graph we get to see 8 different countries having percantage of population age 15-19 no education . Among the 8 countries afghanisthan,pakisthan has the low literacy where as bangladesh has good no of percentage of people having education

```python
df_2020_pop1 = df.query("Year == 2010").query("indicator == 'BAR.NOED.15UP.FE.ZS'")
df_2020_pop.sample(5)
fig = px.pie(
    df_2020_pop1, 
    values='value', names='Country Name',
    title='2010 Barro-lee:Percentage of female population age 15+ with no education'

)
fig.update_traces(textposition='inside', textinfo='percent+label')



fig.show()
```

In the pie chartwe can see female population having age 15+has no education even in female population having no education the afghanisthan stand in the first place where as here  the bangladesh has low amount of female having education.china has more female education when compared to other education
